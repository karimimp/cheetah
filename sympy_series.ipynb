```python
import sympy
sympy.init_printing()
from sympy import I, pi, oo
x = sympy.Symbol("x")
y = sympy.Symbol("y")
z = sympy.Symbol("z")
```

```python
f = sympy.Function("f")(x)
```

```python
sympy.series(f, x)
```

```python
x0 = sympy.Symbol("{x_0}")
```

```python
f.series(x, x0, n=2)
```
