```python
from scipy import interpolate
from numpy import polynomial as P
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def runge(x):
    return 1/(1 + 25 * x**2)
```

```python
x = np.linspace(-1, 1, 11)
y = runge(x)
f_i = interpolate.interp1d(x, y, kind=3) # 3rd order spline
```

```python
xx = np.linspace(-1, 1, 100)
fig, ax = plt.subplots(figsize=(8,4), dpi=100)
ax.plot(xx, runge(xx), 'k', lw=2, label="Rung's function")
ax.plot(x, y, 'ro', label="sample points")
ax.plot(xx, f_i(xx), 'r--', lw=2, label="spline order 3")
ax.legend(loc='best')
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
```

Splines of different orders
---

```python
x = np.array([0,1,2,3,4,5,6,7])
y = np.array([3,4,3.5,2,1,1.5,1.25,0.9])
xx = np.linspace(x.min(), x.max(), 100)
fig, ax = plt.subplots(figsize=(8,4), dpi=100)
ax.scatter(x,y)

for n in [1,2,3,5,7]:
    f=interpolate.interp1d(x, y, kind=n)
    ax.plot(xx, f(xx), label="order %d" %n)

ax.legend(loc='best')
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
fig.tight_layout()
```
